**IMPORTING LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris
```

**LOAD THE IRIS DATASET**

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['sepal.width']=iris.sepalwidth
```

**CREATING BAR CHART**

```python
plt.figure(figsize=(7.5, 4.5))
sns.countplot(x='target', data=df, palette='viridis')
plt.xlabel('Species', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Distribution of Iris Species', fontsize=16)
plt.xticks(ticks=range(3), labels=iris.target_names, fontsize=12, rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**CREATING A HISTOGRAM**

```python
sepal_length = df['sepal length (cm)']
sns.set(style="whitegrid", palette="pastel")
plt.figure(figsize=(7.5, 4.5))
plt.hist(sepal_length, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Sepal Length (cm)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Sepal Length in Iris Dataset', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
